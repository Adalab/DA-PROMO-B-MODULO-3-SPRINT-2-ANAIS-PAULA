TV-Y - menores de 6 aos

TV-Y7 - + 7
TV-Y7-FV - + 7 (violencia y emociones fuertes)

TV-G - pblico general
G - general audiences

TV-PG - supervision parental
PG - parental guidance suggested

PG-13 - mayores de 13 aos
TV-14 - + 14

nan


TV-MA - + 18 o con supervisin
R- restricted, no recomendable para menores, aunque se puede ver con supervisin de un adulto
NC-17 - + 18

UR - unrated
NR - non rated

```python
def cambiar_14 (x):
    if x == 'PG-13':
        x = '+14'
        return x
    elif x == 'TV-14':
        x = '+14'
        return x
    else:
        return x
```

```python
df['rating'] = df['rating'].apply(cambiar_14)
df["rating"].unique()
```

```python
def cambiar_sp (x):
    if x == 'TV-PG':
        x = 'SP'
        return x
    elif x == 'PG':
        x = 'SP'
        return x
    else:
        return x
```

```python
df['rating'] = df['rating'].apply(cambiar_sp)
df["rating"].unique()
```
