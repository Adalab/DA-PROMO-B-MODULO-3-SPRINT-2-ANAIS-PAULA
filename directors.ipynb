  > Sacamos los directores por separado

```python
def separar (col):
    try:
        return col.split(',')
    except:
        pass
```

```python
df['director'] = df['director'].apply(separar)
```

```python
df = df.explode('director')
```

```python
def espais (col):
    try:
        return col.strip()
    except:
        pass
```

```python
df['director'] = df['director'].apply(espais)
```

```python
df.head()
```

```python
list(df['director'].unique())
```

```python
df.groupby(['director']).size()
```

```python
df.columns
```

```python
#top_direct = df.groupby('director')
#top_direct.head(5)
```

```python
top_direct = df.groupby('director').size().reset_index(name = 'count').sort_values(by= 'count', ascending=False)
top_direct.head(10)
```

```python
top_10 = top_direct.iloc[:10,:]
```

```python
top_10['director'].unique()
```

```python
df = df[df['director'].isin(top_10['director'].unique())]
```

```python
df.to_csv('directors.csv')
```
