 > Hacemos lo mismo con countries

```python
dfc = df2.copy()
```

```python
def separar (col):
    try:
        return col.split(',')
    except:
        pass
```

```python
dfc['country'] = dfc['country'].apply(separar)
```

```python
dfc = dfc.explode('country')
```

```python
top_country = dfc.groupby('country').size().reset_index(name = 'count').sort_values(by= 'count', ascending=False)
top_country.head(10)
```

```python
dfc = dfc.reset_index().drop('index', axis=1)
```

```python
dfc
```

```python
dfc.to_csv('countries.csv')
```
