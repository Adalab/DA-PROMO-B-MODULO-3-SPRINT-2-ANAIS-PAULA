  > Sacamos los directores por separado

```python
def separar (col):
    try:
        return col.split(',')
    except:
        pass
```

```python
df['director'] = df['director'].apply(separar)
```

```python
df = df.explode('director')
```

```python
def espais (col):
    try:
        return col.strip()
    except:
        pass
```

```python
df['director'] = df['director'].apply(espais)
```

```python
df.head()
```

  > Sacamos los directores por separado

```python
df_directors = df_us.copy()
```

```python
list(df_directors['director'].unique())
```

```python
df_directors.groupby(['director']).size()
```

```python
top_direct_us = df_directors.groupby('director').size().reset_index(name = 'count').sort_values(by= 'count', ascending=False)
top_direct_us.head(10)
```

```python
top_5_us = top_direct_us.iloc[:5,:]
```

```python
top_5_us['director'].unique()
```

```python
df_directors = df_directors[df_directors['director'].isin(top_5_us['director'].unique())]
```

```python
df_directors.to_csv('directors_us.csv')
```
